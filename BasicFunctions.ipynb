```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

from IPython.display import display
```

```python
def plotTimeSeries(node_id):

    df = pd.read_csv(node_id+".csv")
    df['Date'] = pd.to_datetime(df['Date'])
    df

    # Plot the time series
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.plot(df['Date'], df['Value'], marker='o', linestyle='-')

    plt.title('Time Series Plot')
    plt.xlabel('Date')
    plt.ylabel('Value')
    plt.grid(True)

    return plt.show()
```

```python
def removeErroneousData(node_id):

    df = pd.read_csv(node_id+".csv")
    df['Date'] = pd.to_datetime(df['Date'])
    df

    #Removes minor negative values
    for x in df.index: 
        if df.loc[x, "Value"] < 0:
            if df.loc[x, "Value"] > -1 :
                df.loc[x, "Value"] = 0


    #Finds average of values
    mean = df["Value"].mean()

    #Finds standard deviation
    std = df["Value"].std()

    #Fills normVals vector with values that are not outliers
    normVals = []
    
    for x in df.index:
        val = df.loc[x,"Value"]
    
        dist = abs(mean - val)

        stdDist = abs(dist/std)
    
        if (stdDist <= 2):
            normVals = df.index
            
    for x in df.index:
        if df.loc[x, "Value"] > 0:
            break
        else: 
            df.loc[x, "Value"] = abs(df.loc[x, "Value"])

    return df

    #Plot the time series
    # fig, ax = plt.subplots(figsize=(10, 6))
    # ax.plot(df['Date'], df['Value'], marker='o', linestyle='-')

    # plt.title('Time Series Plot')
    # plt.xlabel('Date')
    # plt.ylabel('Value')
    # plt.grid(True)

    # return plt.show()
```

```python
df = pd.read_csv("001B.csv")
df['Date'] = pd.to_datetime(df['Date'])
df 
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(df['Date'], df['Value'], marker='o', linestyle='-')

df2 = removeErroneousData("001B")
ax.plot(df2['Date'], df2['Value'], marker='o', linestyle='-', color = 'orange')

plt.title('Time Series Plot')
plt.xlabel('Date')
plt.ylabel('Value')
plt.grid(True)

plt.show()
```
